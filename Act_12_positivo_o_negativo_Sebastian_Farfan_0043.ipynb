## Las variables

1. NUM: Valor para determinar su signo
2. R: Resultado del signo del valor

## Formulas

* NUM >= 0
* Si el número es mayor o igual a 0 $R = \text{"POSITIVO"}$
* Si el número es menor a 0 $R = \text{"NEGATIVO"}$

## Diagrama de flujo en LucidChart
<img src="https://raw.githubusercontent.com/sebastianfarfan128/Act-12-problema-3.2-pag-53/refs/heads/main/Diagrama%20de%20flujo%20positivo%20o%20negativo%20Act%2012.png" width="350" height="400" alt="Texto descriptivo" />

```python
# Programa para determinar si un número es positivo o negativo
# Basado en la lógica de comparación simple
# Datos de entrada: Se lee el número (NUM)

NUM = float(input("Ingrese el número: "))

if NUM >= 0:
    R = "POSITIVO"
else:
    R = "NEGATIVO"

# Mostrar el resultado final
print("El número es:", R)

print("Fin del proceso")
print("Programa realizado por Sebastian Farfan NL:16")
```
